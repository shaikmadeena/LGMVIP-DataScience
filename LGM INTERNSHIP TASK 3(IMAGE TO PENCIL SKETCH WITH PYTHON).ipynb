# IMAGE TO PENCIL SKETCH WITH PYTHON

```python
import cv2
from PIL import Image
from IPython.display import display
```

## READ  IMAGE

```python
image=cv2.imread("Desktop\dog.jpeg")
```

## DISPLAY IMAGE 

```python
image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
display(Image.fromarray(image))
```

## CONVERT IMAGE INTO GRAYSCALE IMAGE 

```python
gray_image=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
display(Image.fromarray(gray_image))
```

## CONVERT GRAYSCALE IMAGE INTO NEGATIVE IMAGE

```python
invert_image=cv2.bitwise_not(gray_image)
display(Image.fromarray(invert_image))
```

## CONVERT NEGATIVE IMAGE TO BLURRY IMAGE

```python
blur_image=cv2.GaussianBlur(invert_image, (21,21),0)
display(Image.fromarray(blur_image))
```

## CREATE PENCIL SKETCH BY MIXING GRAYSCALE IMAGE WITH INVERTED BLURRY IMAGE

```python
invert_blur = cv2.bitwise_not(blur_image)
Pencil_Sketch = cv2.divide(gray_image,invert_blur,scale=252.0)
display(Image.fromarray(Pencil_Sketch))
```
